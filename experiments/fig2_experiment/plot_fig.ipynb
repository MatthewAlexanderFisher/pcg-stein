```python
import os
import jax, jax.numpy as jnp
import numpy as np
import matplotlib.pyplot as plt
import yaml
import pandas as pd

from pcg_stein.registry import PRECON_REGISTRY
from pcg_stein.util import make_main_plot

import seaborn as sns


import matplotlib as mpl
mpl.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
})

os.getcwd()
```

#### Load Data and Define Metrics

```python
cg_results = pd.read_csv("results/fig2_cg.csv")
pcg_results = pd.read_csv("results/fig2_pcg.csv")
```

### Create Figures and Save

```python
cg_vals = cg_results["value"].to_numpy()
pcg_vals = pcg_results["value"].to_numpy()

plt.figure(figsize=(8,4))
plt.plot(cg_vals,  label='CG',  linewidth=2)
plt.plot(pcg_vals, label='PCG', linewidth=2, linestyle='--')
plt.xlabel(r'Number of iterations $m$', fontsize=14)
plt.ylabel(r'$\ln \sigma(\mathbf{w}_m)$', fontsize=14)
plt.yscale('log')        # optional: log-scale if values drop exponentially
plt.legend(loc='upper right', fontsize=14)
plt.grid(True, which='both', ls=':')
plt.tight_layout()
plt.savefig(
    "figs/fig2.pdf",  
    format="pdf",     
    bbox_inches="tight",  
    pad_inches=0          
)
```
